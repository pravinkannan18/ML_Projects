```python
from sklearn.preprocessing import LabelEncoder
label_encode = LabelEncoder()
label = label_encode.fit_transform(house_price.date)
label1 = label_encode.fit_transform(house_price.street)
label2 = label_encode.fit_transform(house_price.city)
label3 = label_encode.fit_transform(house_price.statezip)
label4 = label_encode.fit_transform(house_price.country)
```

```python
house_price['date'] = label
house_price['street'] = label1
house_price['city'] = label2
house_price['statezip'] = label3
house_price['country'] = label4
```

```python
house_price.info()
```

```python
x=house_price.drop(columns=['sqft_above','sqft_basement','view','condition'],axis=1)
y=house_price['price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
import tensorflow as tf

# Step 5: Implement the model using TensorFlow
tf_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(x_train.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compile the model
tf_model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
tf_model.fit(x_train, y_train, epochs=50, batch_size=32, verbose=0)

# Evaluate the model
mse_tf = tf_model.evaluate(x_test, y_test, verbose=0)
print("TensorFlow Model Mean Squared Error:", mse_tf)
```
